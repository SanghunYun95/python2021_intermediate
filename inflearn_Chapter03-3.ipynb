```python
class Car():
    
    """
    Car class
    Author: Yun
    Date: 2021.01.29
    Description: Class, Static, Instance Method
    """
    
    # 클래스 변수 (모든 인스턴스 공유)
    price_per_raise = 1.0
    
    # 인스턴스 메소드의 첫번째 매개변수는 self
    def __init__(self, company, details):
        self._company = company # 인스턴스 변수에는 _를 붙여서 클래스변수와 구분짓는다
        self._details = details
        
    def __str__(self):
        return 'str: {} - {}'.format(self._company, self._details)
    
    def __repr__(self):
        return 'repr: {} - {}'.format(self._company, self._details)
    
    def detail_info(self):
        print('Current ID: {}'.format(id(self)))
        print('Car Detail Info: {} {}'.format(self._company, self._details.get('price')))
    
    def get_price(self):
        return 'Before Car Price -> company : {}, price : {}'.format(self._company, self._details.get('price'))
    
    def get_price_calc(self):
        return 'After Car Price -> company : {}, price : {}'.format(self._company, self._details.get('price') *
                                                                    Car.price_per_raise)  
    # Class Method
    @classmethod
    def raise_price(cls, per):
        if per <= 1:
            print('Please Enter 1 or Greater')
            return
        cls.price_per_raise = per
        print('Succeed! price increased')
        
    # Static Method
    # Class Method는 cls를 파라미터로 받기 때문에, Class 변수에 접근가능하며 State에 대한 변형이 가능하다
    # Static Method는 파라미터를 받지 않기 때문에, Class에 접근이 불가능하며 유연하게 작성할 수 있다.
    # 클래스 공통은 아니지만, 각 인스턴스를 아우르는 포괄적이고 유연한 메소드를 작성할 때
    # 자주 쓰이지는 않지만 인스턴스 메소드로 작성하기도 애매한 경우 사용한다.
    # Static Method를 굳이 사용해야 하는가에 대해서 여러 의견이 있다.
    # 따로 빼서 쓰긴 애매하고 클래스에 같이 묶여 있으면 좋겠을 때 사용한다.
    @staticmethod
    def is_bmw(inst):
        if inst._company == 'BMW':
            return 'OK! This car is {}'.format(inst._company)
        return 'Sorry this car is not BMW'
            
car1 = Car('Ferrari', {'color': 'Red', 'horsepower': 400, 'price': 8000})
car2 = Car('BMW', {'color': 'Silver', 'horsepower': 270, 'price': 5000})

# 이렇게 직접 접근하는것은 좋지 않다
print(car1._details.get('price'))

# 메소드를 통한 가격정보
print(car1.get_price())

# 가격 인상 (클래스 메소드 미사용)
Car.price_per_raise = 1.4
print(car1.get_price_calc())

# 가격 인상 (클래스 메소드 사용)
Car.raise_price(1.6)
print(car1.get_price_calc())

# 클래스로 호출도 가능하고 인스턴스로 호출도 가능하다 
print(car1.is_bmw(car1))
print(car2.is_bmw(car2))
print(Car.is_bmw(car1))
print(Car.is_bmw(car2))
```
